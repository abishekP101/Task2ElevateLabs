# feature Engineering

```python
data['Ticket'].value_counts()
```

```python
data[data['Ticket']=='CA. 2343']
```

```python
data['individual_fare'] = data['Fare']/(data['SibSp']+data['Parch']+1)
```

```python
data[['individual_fare' , 'Fare']].describe()
```

```python
data['individual_fare'].plot(kind='box')
```

```python
data['Family_size'] = data['SibSp']+data['Parch']+1
```

```python
def transform_data(num):
    if num == 1:
        return 'alone'
    elif num>1 and num<5:
        return 'small'
    else:
        return 'large'
```

```python
data['family_type'] = data['Family_size'].apply(transform_data)
```

```python
data.head(5)
```

```python
pd.crosstab(data['Survived'] , data['family_type'] , normalize='columns')*100
```

```python
data['Title']= data['Name'].str.split(',').str.get(1).str.split(' ').str.get(1)
```

```python
data['Title']
```

```python
data['Title'].value_counts()
```

```python
data['Title'] = data['Title'].str.replace('Rev.','other')
data['Title'] = data['Title'].str.replace('Dr.','other')
data['Title'] = data['Title'].str.replace('Col.','other')
data['Title'] = data['Title'].str.replace('Jonkheer.','other')
data['Title'] = data['Title'].str.replace('Capt.','other')
data['Title'] = data['Title'].str.replace('the','other')
data['Title'] = data['Title'].str.replace('Major.','other')
data['Title'] = data['Title'].str.replace('Lady.','other')
data['Title'] = data['Title'].str.replace('Mme.','other')
data['Title'] = data['Title'].str.replace('Mlle.','other')
data['Title'] = data['Title'].str.replace('Don.','other')
```

```python
temp_df = data[data['Title'].isin(['Mr.','Miss.' , 'Master.' , 'other'])]
```

```python
pd.crosstab(temp_df['Survived'] , temp_df['Title'], normalize='columns')*100
```
